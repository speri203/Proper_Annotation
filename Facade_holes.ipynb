```python
import json
import os
```

```python
datasets = []
for i, items in enumerate(os.listdir('Datasets')):
    datasets.append(items)
    print("Index: {} Name: {}".format(i, items))
```

```python
selected_ds = input("Which dataset would you like to use (input index number)")
print(datasets[int(selected_ds)])
ann_folder = 'Datasets\\' + datasets[int(selected_ds)]
```

```python
## Open up json file and save all contents of windows, ground, door
os.listdir()
for files in os.listdir(ann_folder + '\\ann'):
    if(files.endswith('.json')):
        #print(files)
        with open(ann_folder + '\\ann\\' + files) as json_file:
            ## Will hold all the coordinates of the specific object
            windows = []
            doors = []
            hvac = []
            content = json.load(json_file)
            for entry in content['objects']:
                if(entry['classTitle'] == 'Window'):
                    #print(entry['points']['exterior'])
                    windows.append(entry['points']['exterior'])
                    print("{} {}".format(files, len(windows)))
                if(entry['classTitle'] == 'Door'):
                    #print(entry['points']['exterior'])
                    doors.append(entry['points']['exterior'])
                if(entry['classTitle'] == 'HVAC'):
                    #print(entry['points']['exterior'])
                    hvac.append(entry['points']['exterior'])
            #print("Filename : {} Window coord: {} Doors: {}".format(files, windows, doors))
            for entry in content['objects']:
                test = []
                if(entry['classTitle'] == 'Facade' or entry['classTitle'] == 'Facades'):
                    entry['tags'] = [{"name": "Facade"}]
                    if(len(windows) > 0):
                        entry['points']['interior'] = windows
                    if(len(doors) > 0):
                        entry['points']['interior'] += doors
                    if(len(hvac) > 0):
                        entry['points']['interior'] += hvac
                if(entry['classTitle'] == 'Roof'):
                    #entry['tags'] = [{"name": "Roof"}]
                    if(len(hvac) > 0):
                        entry['points']['interior'] += hvac
                if(entry['classTitle'] == 'HVAC'):
                    entry['tags'] = [{"name": "HVAC"}]
                if(entry['classTitle'] == 'Window'):
                    entry['tags'] = [{"name": "Window"}]
                if(entry['classTitle'] == 'Door'):
                    entry['tags'] = [{"name": "Door"}]
                if(entry['classTitle'] == 'Sky'):
                    entry['tags'] = [{"name": "Sky"}]
                if(entry['classTitle'] == 'Ground'):
                    entry['tags'] = [{"name": "Ground"}]
                    
        with open(ann_folder + '\\proper_ann\\' + files, 'w') as writing:
            json.dump(content, writing)

print("Done")
```
